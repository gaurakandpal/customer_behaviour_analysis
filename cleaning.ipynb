```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isna().sum()
```

```python
labels=['young adult','adult','middle age','senior']
df['Age Group']=pd.qcut(df['Age'],q=4,labels=labels)
```

```python
fre_mapping={
    'Fortnightly': 14,
    'Weekly':7, 
    'Annually':365, 
    'Quarterly':90, 
    'Bi-Weekly':14,
    'Monthly':30, 
    'Every 3 Months':90
}
df['frequency_days']=df['Frequency of Purchases'].map(fre_mapping)
```
